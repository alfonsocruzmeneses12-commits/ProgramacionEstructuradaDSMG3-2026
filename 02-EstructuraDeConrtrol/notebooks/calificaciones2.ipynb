# Programa de calificaiones
Este programa calcula el promedio de 3 calificaciones si un alumno **"Alumno"** o **"No aprobado"**

## Definicion de Variables


```python
calif1 = 0.0
calif2 = 0.0
calif3 =0.0
promedio =0.0
```

## Entrada de Datos

```python
calif1 = float (input ("introducce la calificacion 1"))
calif2 = float (input ("introducce la calificacion 2"))
calif3 = float (input ("introducce la calificacion 3"))
```

## Prceso Calculo del promedio

```python
promedio = (calif1 + calif2 + calif3) / 3
```

## Comparacion con una estructura selecctiva doble 

```python
if promedio >= 7.0:
 print("Aprobado")
else:
 print ("No aprobado")
```
